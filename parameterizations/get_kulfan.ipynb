```python
import aerosandbox as asb
import numpy as np
import pandas as pd
from aerosandbox.geometry.airfoil.airfoil_families import get_kulfan_parameters
from aerosandbox.geometry.airfoil.airfoil_families import get_kulfan_coordinates
```

```python
df = pd.read_excel(r"C:\Users\rafae\OneDrive\Área de Trabalho\TCC\airfoil_data_to_diff.xlsx")
```

```python
airfoils_list = df['AIRFOIL_NAME'].tolist()
```

```python
# Create an empty list to store the results
data_list = []

for airfoil_name in airfoils_list:
    af = asb.Airfoil(airfoil_name)
    params = get_kulfan_parameters(coordinates=af.coordinates)
    
    # Split the arrays for 'lower_weights' and 'upper_weights' into individual items
    for i, value in enumerate(params['lower_weights'], 1):
        params[f'lower_weight{i}'] = value

    for i, value in enumerate(params['upper_weights'], 1):
        params[f'upper_weight{i}'] = value

    # Remove the original 'lower_weights' and 'upper_weights' keys
    del params['lower_weights']
    del params['upper_weights']

    # Add the airfoil name to the dictionary
    params['Airfoil Name'] = airfoil_name

    data_list.append(params)

# Create a DataFrame from the list of dictionaries
results_df = pd.DataFrame(data_list)

# Save the DataFrame to an Excel file
results_df.to_excel('airfoils_kulfan.xlsx', index=False)
```
