```python
import aerosandbox as asb
import numpy as np
import matplotlib.pyplot as plt

# Create the airfoil object
af2 = asb.Airfoil("NACA0024")

# Generate cosine spaced distribution
num_points = 150
theta = np.linspace(0, np.pi, num_points)
x_cosine = (1 - np.cos(theta)) / 2

# Compute y-values for upper and lower surfaces
y_upper2 = af2.local_camber(x_cosine) + af2.local_thickness(x_cosine) / 2
y_lower2 = af2.local_camber(x_cosine) - af2.local_thickness(x_cosine) / 2

# Plotting using scatter
plt.scatter(x_cosine, y_upper2, s=2, label="Superior")
plt.scatter(x_cosine, y_lower2, s=2, label="Inferior")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Perfil NACA0024 com espaçamento cossenoidal")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
